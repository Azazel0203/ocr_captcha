# Data Loading

````python
# Define the directory containing the images
data_dir = Path("/content/train_images_mlware/train_images")
# Fetch paths of all PNG images in the directory and sort them
images = sorted(list(map(str, list(data_dir.glob("*.png")))))
# Extract image IDs from the image paths
ids = [img.split(os.path.sep)[-1] for img in images]
# Read the CSV file containing image labels into a DataFrame
data_sheet = pd.read_csv("/content/train-labels_mlware.csv")
# Create a dictionary mapping image IDs to their respective labels
data_dict = {data_sheet['image'][i]: data_sheet['text'][i] for i in range(25000)}
# Retrieve labels for each image ID
labels = [data_dict[item] for item in ids]
# Extract unique characters present in all labels
characters = set(char for label in labels for char in label)
# Sort the characters alphabetically
characters = sorted(list(characters))
# Print information about the dataset
print("Number of images found: ", len(images))
print("Number of labels found: ", len(labels))
print("Number of unique characters: ", len(characters))
print("Characters present: ", characters)
# Find the maximum length of labels
max_length = max([len(label) for label in labels])
print(f"max_length -> {max_length}")
# Find the minimum length of labels
min_length = min([len(label) for label in labels])
print(f"min_length -> {min_length}")
````

creating the mappings from charachter to number and vice versa

````python
# Mapping characters to integers
char_to_num = layers.StringLookup(vocabulary=list(characters), mask_token=None)
# Mapping integers back to original characters
num_to_char = layers.StringLookup(
    vocabulary=char_to_num.get_vocabulary(), mask_token=None, invert=True
)
````

Defining the `split_data` function to split the data into training and validation sets

````python
def split_data(images, labels, train_size=0.8, shuffle=True):
    # 1. Get the total size of the dataset
    size = len(images)
    # 2. Make an indices array and shuffle it, if required
    indices = np.arange(size)
    if shuffle:
        np.random.shuffle(indices)
    # 3. Get the size of training samples
    train_samples = int(size * train_size)
    # 4. Split data into training and validation sets
    x_train, y_train = images[indices[:train_samples]], labels[indices[:train_samples]]
    x_valid, y_valid = images[indices[train_samples:]], labels[indices[train_samples:]]
    return x_train, x_valid, y_train, y_valid
````

Creating the training and validation splits

````python
x_train, x_valid, y_train, y_valid = split_data(np.array(images), np.array(labels))
````

Creating the function `encode_single_sample` to encode an image along with its label and return

````python
def encode_single_sample(img_path, label):
    # 1. Read image
    img = tf.io.read_file(img_path)
    # 2. Decode and convert to grayscale
    img = tf.io.decode_png(img, channels=1)
    # 3. Convert to float32 in [0, 1] range
    img = tf.image.convert_image_dtype(img, tf.float32)
    # 4. Resize to the desired size
    img = tf.image.resize(img, [img_height, img_width])
    # 5. Transpose the image because we want the time
    # dimension to correspond to the width of the image.
    img = tf.transpose(img, perm=[1, 0, 2])
    # heightxwidth -> widthxheight
    # 6. Map the characters in label to numbers
    label = char_to_num(tf.strings.unicode_split(label, input_encoding="UTF-8"))
    # 7. Return a dict as our model is expecting two inputs
    return {"image": img, "label": label}
````

Defining the optimum hyper parameters of batch_size, img_height and img_width

````python
batch_size = 32
img_height = 100
img_width = 200
````

````python
"""
1. convert to tensors
2. use the encode_single_sample to encode the image and the labels
3. the number of parallel threads is decided automatically by "tf.data.AUTOTUNE" for efficiency
4. make it into a batch of "batch_size"
5. the prefetch loads the images and labels into the buffer to optimize and smoothen the training
"""
train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
train_dataset = (
    train_dataset.map(encode_single_sample, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(batch_size)
    .prefetch(buffer_size=tf.data.AUTOTUNE)
)

validation_dataset = tf.data.Dataset.from_tensor_slices((x_valid, y_valid))
validation_dataset = (
    validation_dataset.map(encode_single_sample, num_parallel_calls=tf.data.AUTOTUNE)
    .batch(batch_size)
    .prefetch(buffer_size=tf.data.AUTOTUNE)
)
````

# Model Building

`ctc_label_dense_to_sparse`

---

This function is designed to convert dense label representations to sparse tensors.

1. **Calculate Shapes**:
   - The function starts by computing the shape of the labels tensor (`label_shape`), determining the number of batches (`num_batches`), and identifying the maximum number of labels (`max_num_labels`).

2. **Define Inner Function**:
   - An inner function named `range_less_than(old_input, current_input)` is defined. This function generates a boolean mask indicating which parts of the label sequence are relevant based on label lengths.

3. **Initialize Dense Mask**:
   - A tensor filled with zeros is initialized and then cast to boolean (`init`). This tensor serves as the initial state for the dense mask.

4. **Generate Dense Mask**:
   - Using `tf.compat.v1.scan()` with the `range_less_than` function, a dense mask is generated iteratively over label lengths. This mask highlights valid parts of the label sequence.

5. **Extract Relevant Indices**:
   - Relevant indices are extracted based on the dense mask and label array. Both batch and label indices are calculated to identify specific elements within the labels tensor.

6. **Create Sparse Tensor**:
   - Values are gathered from the labels tensor according to the calculated indices. Subsequently, a sparse tensor is constructed using `tf.SparseTensor()`.

This function efficiently converts dense label representations into sparse tensors, which is particularly valuable in scenarios like Connectionist Temporal Classification (CTC) where sequences have variable lengths and require sparse representations for efficient processing.

````python
def ctc_label_dense_to_sparse(labels, label_lengths):
    label_shape = tf.shape(labels) # B, T, C
    # print("label_shape:", label_shape)
    num_batches = tf.stack([label_shape[0]])
    # print("num_batches:", num_batches)
    max_num_labels = tf.stack([label_shape[1]])
    # print("max_num_labels:", max_num_labels)
    def range_less_than(old_input, current_input):
        '''
        Creates a boolean mask for the label_lengths we need to pay attention to
        '''
        return tf.expand_dims(
            tf.range(tf.shape(old_input)[1]), 0) < tf.fill(max_num_labels, current_input)
    init = tf.cast(tf.fill([1, label_shape[1]], 0), tf.bool)
    '''
    1. Initialize a tensor filled with zeros with shape 1,T
    2. type_Cast it to bool -> tensor of False
    '''
    # print("init:", init)
    dense_mask = tf.compat.v1.scan(range_less_than, label_lengths, initializer=init, parallel_iterations=1)
    # print("dense_mask:", dense_mask)
    dense_mask = dense_mask[:, 0, :]
    # squeeze the middle dimension
    # print("dense_mask after squeeze:", dense_mask)
    label_array = tf.reshape(tf.tile(tf.range(0, label_shape[1]), num_batches), label_shape)
    # print("label_array:", label_array)
    label_ind = tf.compat.v1.boolean_mask(label_array, dense_mask)
    # print("label_ind:", label_ind)
    batch_array = tf.transpose(tf.reshape(tf.tile(tf.range(0, label_shape[0]), max_num_labels), tf.reverse(label_shape, [0]),))
    # print("batch_array:", batch_array)
    batch_ind = tf.compat.v1.boolean_mask(batch_array, dense_mask)
    # print("batch_ind:", batch_ind)
    indices = tf.transpose(tf.reshape(tf.concat([batch_ind, label_ind], axis=0), [2, -1]))
    # print("indices:", indices)
    vals_sparse = tf.compat.v1.gather_nd(labels, indices)
    # print("vals_sparse:", vals_sparse)
    return tf.SparseTensor(
        tf.cast(indices, tf.int64), vals_sparse, tf.cast(label_shape, tf.int64)
    )
````

`ctc_batch_cost`

---

This function computes the Connectionist Temporal Classification (CTC) loss between the predicted logits (`y_pred`) and the true labels (`y_true`).

1. **Adjust Label and Input Lengths**:
   - The function first converts the label and input lengths to `int32` data type after removing any singleton dimensions from the end of the tensors. These lengths are crucial for computing the CTC loss.

2. **Generate Sparse Labels**:
   - It generates a sparse representation of the true labels using the `ctc_label_dense_to_sparse` function. This converts dense label representations into sparse tensors, which is essential for CTC loss calculation.

3. **Compute Logits**:
   - Logarithms of the predicted logits (`y_pred`) are calculated after transposing the tensor to match the required format for CTC loss computation. A small value (`keras.backend.epsilon()`) is added before taking the logarithm to avoid errors when the input is close to zero.

4. **Calculate CTC Loss**:
   - The CTC loss is computed using `tf.compat.v1.nn.ctc_loss`, with inputs being the logarithm of predicted logits (`y_pred`), sparse labels (`sparse_labels`), and sequence lengths (`input_length`). This function computes the CTC loss for each sample in the batch.

5. **Add Batch Dimension**:
   - Finally, the computed CTC loss is expanded by adding a singleton dimension to represent the batch size. This ensures that the shape of the output matches the expected format.

Overall, this function facilitates the calculation of the CTC loss, which is commonly used in sequence prediction tasks such as speech recognition and handwriting recognition. The CTC loss measures the discrepancy between predicted sequences and true sequences, considering possible alignments and variable-length outputs.

````python
def ctc_batch_cost(y_true, y_pred, input_length, label_length):
    label_length = tf.cast(tf.squeeze(label_length, axis=-1), tf.int32)
    # print(f"label_length: {label_length}")
    '''
    1. removes the dimensions of size 1, here its removing the last dimension
        if the value of last dimension of label_length is 1, it will remove that
    2. casting the label_length to int32
    '''
    input_length = tf.cast(tf.squeeze(input_length, axis=-1), tf.int32)
    # print(f"input_length: {input_length}")
    '''
    This is basically label_length but for y_pred
    '''
    sparse_labels = tf.cast(ctc_label_dense_to_sparse(y_true, label_length), tf.int32)
    # print(f"sparse_labels: {sparse_labels}")
    '''
    Generating a concentrated sparse matrix
    '''
    y_pred = tf.math.log(tf.transpose(y_pred, perm=[1, 0 ,2]) + keras.backend.epsilon())
    # print(f"y_pred: {y_pred}")
    '''
    Add a small value of epsilon before taking log...so as to not take the log of 0 by mistake
    '''
    '''
    returns the actual loss value
    add a singleton dim to the output...so as to represent the batch size
    '''
    return tf.expand_dims(tf.compat.v1.nn.ctc_loss(
        inputs=y_pred, labels=sparse_labels, sequence_length=input_length
        ), 1,
    )
````

`CTCLayer`


---


The `CTCLayer` class serves as a custom layer for computing the Connectionist Temporal Classification (CTC) loss within neural network models. Below is a detailed breakdown of its functionality:

1. **Initialization**:
   - The `__init__` method initializes the layer, inheriting properties from the `layers.Layer` class. It sets the loss function for the layer to `ctc_batch_cost`.

2. **Layer Call**:
   - The `call` method is the heart of the layer's functionality. During training, it computes the CTC loss and adds it to the layer using `self.add_loss()`.
   - It calculates the batch size (`batch_len`), input sequence length (`input_length`), and label sequence length (`label_length`) from the shapes of the true labels (`y_true`) and predicted logits (`y_pred`).
   - Subsequently, it creates tensors for input and label lengths with the same length as the batch size.
   - The CTC loss is computed using the `ctc_batch_cost` function with the provided parameters (`y_true`, `y_pred`, `input_length`, `label_length`).
   - The computed loss is then added to the layer using `self.add_loss()`, ensuring it contributes to the model's total loss during training.

3. **Prediction**:
   - During testing or inference, the layer simply returns the predicted logits (`y_pred`) without any further computation. This design allows seamless integration of the layer in both training and inference phases.

Overall, the `CTCLayer` class encapsulates the logic for CTC loss computation within a custom Keras layer, enhancing the modularity and flexibility of neural network models for sequence prediction tasks.

````python
class CTCLayer(layers.Layer):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.loss_fn = ctc_batch_cost

    def call(self, y_true, y_pred):
        # Compute the training-time loss value and add it
        # to the layer using `self.add_loss()`.
        batch_len = tf.cast(tf.shape(y_true)[0], dtype="int64")
        input_length = tf.cast(tf.shape(y_pred)[1], dtype="int64")
        label_length = tf.cast(tf.shape(y_true)[1], dtype="int64")
        input_length = input_length * tf.ones(shape=(batch_len, 1), dtype="int64")
        label_length = label_length * tf.ones(shape=(batch_len, 1), dtype="int64")
        loss = self.loss_fn(y_true, y_pred, input_length, label_length)
        self.add_loss(loss)
        # At test time, just return the computed predictions
        return y_pred
````

`build_model`


---


This function constructs a convolutional neural network (CNN) model followed by recurrent neural networks (RNNs) for Optical Character Recognition (OCR) task. Here's a detailed breakdown of its implementation:

1. **Input Layers**:
   - `input_img`: Input layer for the image data with shape `(img_width, img_height, 1)`.
   - `labels`: Input layer for the labels data with variable length (represented by `None`).

2. **CNN Layers**:
   - Two convolutional blocks are defined (`conv1` and `conv2`) each consisting of a 2D convolutional layer followed by max-pooling.
   - ReLU activation function is used for the convolutional layers.
   - He normal initialization is applied to the convolutional kernels.
   
3. **Reshaping Layer**:
   - The output from the last convolutional layer is reshaped to prepare for feeding into the RNN layers.

4. **Dense Layer**:
   - A dense layer with ReLU activation and dropout regularization is added to introduce non-linearity and prevent overfitting.

5. **RNN Layers**:
   - Two bidirectional LSTM layers are added to capture temporal dependencies in both forward and backward directions.
   - Each LSTM layer is followed by dropout regularization to mitigate overfitting.

6. **Output Layer**:
   - A dense layer with softmax activation is added to produce the final classification output. The number of units in this layer is set to `(len(char_to_num.get_vocabulary()) + 1)` to accommodate the vocabulary size plus one for the CTC blank token.

7. **CTC Loss Layer**:
   - A custom CTC loss layer named `CTCLayer` is added to the model. It computes the CTC loss between the predicted output and true labels.

8. **Model Compilation**:
   - The model is compiled using the Adam optimizer with a learning rate schedule defined by exponential decay.
   - Other optimizer parameters such as `beta_1` and `beta_2` are set to default values.
   
9. **Learning Rate Schedule**:
   - An exponential decay learning rate schedule is defined to adjust the learning rate during training.
   - The initial learning rate is set to `0.001`, and the decay rate is set to `0.9`.
   - Decay steps and staircase mode are adjusted as needed.

10. **Model Definition**:
   - The model is defined with input layers (`input_img` and `labels`) and output layer (`output`).
   - The model is given the name "ocr_model_v1".

11. **Return**:
   - Finally, the constructed model is returned.

Overall, this function constructs a comprehensive OCR model architecture suitable for recognizing text from images, incorporating both CNNs and RNNs, and utilizing the CTC loss for training.



````python
def build_model():
    # inputs of the model
    input_img = layers.Input(shape=(img_width, img_height, 1), name="image", dtype="float32")
    labels = layers.Input(name="label", shape=(None,), dtype="float32")
    # first conv block
    x = layers.Conv2D(64, (3, 3), activation="relu", kernel_initializer="he_normal", padding="same", name="conv1",)(input_img)
    x = layers.MaxPooling2D((2, 2), name="pool1")(x)
    # 2nd conv block
    x = layers.Conv2D(128, (3, 3), activation="relu", kernel_initializer="he_normal", padding="same", name="conv2",)(x)
    x = layers.MaxPooling2D((2, 2), name="pool2")(x)
    # reshaping just in case
    new_shape = ((img_width//4), (img_height//4) * 128) # each spatial location will be denoted by 64 values
    x = layers.Reshape(target_shape=new_shape, name="reshape")(x)
    x = layers.Dense(128, activation="relu", name="dense1")(x)
    x = layers.Dropout(0.2)(x)
    # RNNs
    x = layers.Bidirectional(layers.LSTM(256, return_sequences=True, dropout=0.25))(x)
    x = layers.Bidirectional(layers.LSTM(128, return_sequences=True, dropout=0.25))(x)
    # output_layer
    x = layers.Dense(len(char_to_num.get_vocabulary()) + 1, activation="softmax", name="dense2")(x)
    output = CTCLayer(name="ctc_loss")(labels, x)
    # define the model
    model = keras.models.Model(
        inputs=[input_img, labels],
        outputs=output,
        name="ocr_model_v1"
    )
    # learning rate
    # Define the initial learning rate and decay rate
    initial_learning_rate = 0.001
    decay_rate = 0.9
    # Define the learning rate schedule
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate,
        decay_steps=1000,  # Adjust decay steps as needed
        decay_rate=decay_rate,
        staircase=True  # Set to True for staircase decay
    )
    beta_1 = 0.9
    beta_2 = 0.999
    # optimizer
    opt = tf.keras.optimizers.Adam(learning_rate=lr_schedule, beta_1=beta_1, beta_2=beta_2)
    # compile the model
    model.compile(optimizer=opt)
    return model
````

Single Sample Encoding Function For Testing data


---



1. **Reading Image**:
   - The function reads the image file located at the specified `img_path` using TensorFlow's `tf.io.read_file` function.

2. **Decoding and Converting to Grayscale**:
   - The image is decoded from PNG format and converted to grayscale using `tf.io.decode_png` with the `channels=1` argument.

3. **Converting to Float32 in [0, 1] Range**:
   - The pixel values of the grayscale image are converted to the `float32` data type and normalized to the range [0, 1] using `tf.image.convert_image_dtype`.

4. **Resizing to Desired Size**:
   - The image is resized to the desired dimensions (`img_height` and `img_width`) using `tf.image.resize`.

5. **Transposing Image**:
   - The image is transposed to change the order of its dimensions. This is performed to ensure that the time dimension (width of the image) corresponds to the first dimension.

6. **Returning as Dictionary**:
   - The function returns a dictionary containing the preprocessed image (`img`) and its associated ID (`ids`). This dictionary format matches the input format expected by the model.



````python
def encode_single_sample_testing(img_path, ids):
    # 1. Read image
    img = tf.io.read_file(img_path)
    # 2. Decode and convert to grayscale
    img = tf.io.decode_png(img, channels=1)
    # 3. Convert to float32 in [0, 1] range
    img = tf.image.convert_image_dtype(img, tf.float32)
    # 4. Resize to the desired size
    img = tf.image.resize(img, [img_height, img_width])
    # 5. Transpose the image because we want the time
    # dimension to correspond to the width of the image.
    img = tf.transpose(img, perm=[1, 0, 2])
    # heightxwidth -> widthxheight
    # 6. Map the characters in label to numbers
    # label = char_to_num(tf.strings.unicode_split(label, input_encoding="UTF-8"))
    # 7. Return a dict as our model is expecting two inputs
    return {"image": img, "ids": ids}
````

CTC Decoding Functions


---

1. `ctc_decode(y_pred, input_length, greedy, beam_width=100, top_paths=1)`

- This function performs CTC decoding on the predicted output probabilities (`y_pred`) generated by a neural network model.
- **Parameters**:
  - `y_pred`: Predicted output probabilities from the model.
  - `input_length`: Length of each input sequence.
  - `greedy`: Boolean flag indicating whether to use greedy decoding or beam search decoding.
  - `beam_width`: Beam width parameter for beam search decoding (default value is 100).
  - `top_paths`: Number of top paths to consider in beam search decoding (default value is 1).
- **Returns**:
  - `decoded_dense`: Decoded output sequences in dense format.
  - `log_prob`: Log probability of the decoded sequences.

2. `decode_batch_predictions(pred)`

- This function decodes the batch predictions generated by a neural network model into human-readable text.
- It first determines the input length for each prediction based on the shape of the prediction array.
- Then, it calls the `ctc_decode` function with greedy decoding enabled to obtain the decoded output sequences.
- Finally, it converts the decoded sequences into text format using the character mapping (`num_to_char`) and returns a list of decoded texts.
- **Parameters**:
  - `pred`: Batch predictions generated by the model.
- **Returns**:
  - `output_text`: List of decoded texts corresponding to the batch predictions.



````python
def ctc_decode(y_pred, input_length, greedy, beam_width=100, top_paths=1):
    input_shape = tf.shape(y_pred)
    num_samples, num_steps = input_shape[0], input_shape[1]
    y_pred = tf.math.log(tf.transpose(y_pred, perm=[1, 0, 2]) + keras.backend.epsilon())
    input_length = tf.cast(input_length, tf.int32)
    # print (f"y_pred = {tf.shape(y_pred)}, input_length = {tf.shape(input_length)}")
    if greedy:
        (decoded, log_prob) = tf.nn.ctc_greedy_decoder(
            inputs=y_pred, sequence_length=input_length
        )
    else:
        (decoded, log_prob) = tf.compat.v1.nn.ctc_beam_search_decoder(
            inputs=y_pred,
            sequence_length=input_length,
            beam_width=beam_width,
            top_paths=top_paths,
        )
    decoded_dense = []
    for st in decoded:
        st = tf.SparseTensor(st.indices, st.values, (num_samples, num_steps))
        decoded_dense.append(tf.sparse.to_dense(sp_input=st, default_value=-1))
    return (decoded_dense, log_prob)
def decode_batch_predictions(pred):
    input_len = np.ones(pred.shape[0]) * pred.shape[1]
    # results = ctc_decode(pred, input_length=input_len, greedy=True)[0][0][:, :max_length]
    results = ctc_decode(pred, input_length=input_len, greedy=True)[0][0][
        :, :max_length
    ]
    output_text = []
    for res in results:
        res = tf.strings.reduce_join(num_to_char(res)).numpy().decode("utf-8")
        output_text.append(res)
    return output_text
````
